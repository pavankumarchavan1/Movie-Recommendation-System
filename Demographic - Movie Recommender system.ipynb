```python
def convert(obj):
    l2 = []
    for i in obj:
        l2.append(i['name'])
    return l2
```

```python
def convert(obj):
    l2 = []
    for i in ast.literal_eval(obj):
        l2.append(i['name'])
    return l2
```

```python
movies2['genres'] = movies2['genres'].apply(convert)
```

```python
def convert3(obj):
    m = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            m.append(i['name'])
            counter+=1
        else:
            break
    return m
```

```python
movies2['cast'] = movies2['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies2['crew'] = movies2['crew'].apply(fetch_director)
```

```python
C = data['vote_average'].mean()
```

```python
C
```

```python
m = data['vote_count'].quantile(0.75) # for a movie to feature in the charts, it must have more votes than at least 70% of the movies in the list.
```

```python
m #for a movie to feature, it must have minimum of 737 votes.
```

```python
#Lets filter the movies with this criteria.
```

```python
qualified_movies = movies2.copy().loc[movies2['vote_count'] >= m] 
```

```python
qualified_movies
```

```python
qualified_movies.head(1)
```

```python
qualified_movies.shape
```

```python
def weighted_rating(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    # Calculation based on the IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#greatest hits/classic movies - recommendation based on highest ratings calculated using total votes

#we can use the average rating provided by the voters to rank, however this has a major flaw in terms of skewness due to 
#movies with very low number of votes. For example, movie with 8.5 rating with only 3 votes may not be better than movie 
#with 7.8 rating with 100 votes.

#take the weighted rating approach followed by IMDB

#Weighted Rating (WR) = [(v/(v+m))*R] + [(m/(v+m))*C]

#v -> number of votes for the movie
#m -> minimum votes required to be listed in the chart
#R -> average rating of the movie
#C -> mean vote across the whole report
```

```python
qualified_movies['score'] = qualified_movies.apply(weighted_rating, axis=1)
```

```python
qualified_movies = qualified_movies.sort_values('score', ascending=False)
```

```python
qualified_movies[['title', 'vote_count', 'vote_average', 'score']].head(10) #top 10 highest rated movies
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,8))

plt.barh(qualified_movies['title'].head(10),qualified_movies['score'].head(10), align='center',color='skyblue')
plt.gca().invert_yaxis()
plt.xlabel("Weighted Score", weight='bold')
plt.title("Best Rated Movies",weight='bold')
plt.show()    #greatest hits/classic movies - recommendation based on highest ratings calculated using total votes
```
